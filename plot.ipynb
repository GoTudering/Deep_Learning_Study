```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(a,t):
  return (a-30)*(1.5**t)+30
def g(a,t):
  return (a-55)*(1.21**t)+55
```

```python
t1=np.arange(0,5,0.01)
t2=np.arange(0,5,0.01)
plt.plot(t1,f(40,t1),t2,g(60,t2))
plt.ylim(43,80)
#plt.axvline(x=3,color='r',linestyle='--')
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
s1 = np.linspace(30, 100, 71)
s2 = np.linspace(0, 5, 10)
X, Y = np.meshgrid(s1, s2)

ax.plot_surface(X,Y,g(X,Y))
ax.set_zlim(-10, 10)

plt.tight_layout()
plt.show()
```
